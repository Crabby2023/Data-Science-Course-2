```python
import yfinance as yf

symbol = 'TSLA'

tesla = yf.Ticker(symbol)


tesla_data = tesla.history(period="max") 

```

```python
print(tesla_data.tail())
```

```python
import yfinance as yf

gme_data = yf.Ticker("GME").history(period="max")

gme_data.reset_index(inplace=True)

print(gme_data.head())
```

```python
print(gme_data.tail())
```

```python
tesla_data.reset_index(inplace=True)
```

```python
print(tesla_data.columns)
```

```python
import matplotlib.pyplot as plt

def make_graph(tesla_data, title):
    plt.figure(figsize=(10,6))
    plt.plot(tesla_data['Date'], tesla_data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.show()


make_graph(tesla_data, "Tesla Stock Prices Over Time")

```
